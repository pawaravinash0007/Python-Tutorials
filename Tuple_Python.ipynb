```python
list1=[10,20,30,40]
print(list1,"its data type is",type(list1))
a=tuple(list1)
print(a,"Its data type is",type(a))
```

## Zip multiple list as argument together 

```python
city_list=['Aurangabad','Daman']
post_name=['Senior','Faculty']
print(city_list)
print(post_name)
```

```python
zip(city_list,post_name)
```

```python
list(zip(city_list,post_name))
```

```python
tuple(zip(city_list,post_name))
```
