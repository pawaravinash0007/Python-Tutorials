##  Logical Statement AND
#Returns True if both statements are true

```python
a = 11
b= 8
c=a>=10 and b<=10  # &
print(c)
```

# Logical OR
# Returns True if one of the statements is true

```python
a = 11
b= 80
c=a>=10 or b<=10  # |
print(c)
```
